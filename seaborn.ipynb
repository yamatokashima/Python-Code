### Load Seaborn and Pandas Libaries, along with the dataset called "mpg" using the `load.dataset` command

```python
# Import Library
import seaborn as sns
import pandas as pd

# Load the example mpg.csv dataset
# Reference for this command is: https://seaborn.pydata.org/generated/seaborn.load_dataset.html
tips = sns.load_dataset('mpg')
```

```python
# Take a moment become familiar with the dataset using your favorite Python Pandas exploratory commands.
mpg.head()
```

```python
mpg.describe()
```

```python
mpg.info()
```

### Set the Theme

Use this reference link for set_theme selections: https://seaborn.pydata.org/generated/seaborn.set_theme.html <br>
For the example below using the style of white will match the gallery example.
<br>


```python
# Set Theme
sns.set_theme(context='notebook', style='darkgrid', palette='deep', font='sans-serif', font_scale=1, color_codes=True, rc=None)
```

Use `sns.set_theme(style="white")`

 Use the Seaborn chart type command called `relplot`.  Adjust the chart parameters for hue, size, sizes, alpha, palette, height, and data.<br> Details for the parameters can be found at https://seaborn.pydata.org/generated/seaborn.relplot.html

```python
# sns.relplot()
sns.relplot(x=mpg['horsepower'],y=mpg['mpg'],hue=mpg['origin'],size=mpg['weight'],alpha=0.5,palette='husl',height=4)
```

```python
# Change the variables and create a different version of the same chart!
sns.relplot(x=mpg['model_year'],y=mpg['acceleration'],hue=mpg['horsepower'],size=mpg['cylinders'],alpha=0.5,palette='husl',height=4)
```

### For the next two visualizations, load the provided "tips" dataset. 

Seaborn provides the `load_dataset` command to accomplish this task.

```python
# Load the example tips dataset
tips = sns.load_dataset('tips')
```

```python
# Take a moment become familiar with the dataset using your favorite Python Pandas exploratory commands.
tips.describe()
```

```python
tips.info()
```

```python
tips.head()
```

### Set the Theme

Previously in this notebook, the theme was set to `white` let's try another one.
For the example below using the style of `whitegrid` will match the example.

```python
# Set a new Theme
sns.set_theme(context='paper', style='whitegrid', palette='deep', font='sans-serif', font_scale=1, color_codes=True, rc=None)
```

Use `sns.set_theme(style="whitegrid")`

 Use the Seaborn chart type command called `violinplot`.  Adjust the chart parameters to match the illustration above.<br> Split the Violin Plot with `split` to split the violin.  Details for the parameters can be found at https://seaborn.pydata.org/generated/seaborn.violinplot.html

```python
# Create the Violin Plot using violinplot chart type with despin to split the violin.
sns.violinplot(x=tips['day'],y=tips['total_bill'],hue=tips['smoker'],split=True)
```

```python
# Change a component of the parameters or selected dimensions and recreate.
sns.violinplot(x=tips['tip'],y=tips['sex'],hue=tips['time'],split=True)
```

### Reset the Theme

Let's explore one more theme. For the example below using the style of `darkgrid` will match the upcoming example.

```python
# Assign the darkgrid theme
sns.set_theme(context='paper', style='darkgrid', palette='deep', font='sans-serif', font_scale=1, color_codes=True, rc=None)
```

---

Create a Regression Seaborn chart type command called `jointplot`.  Adjust the chart parameters to match the illustration above.<br> Apply the following settings for xlim and ylim: `xlim=(0, 60), ylim=(0, 12)`.  Details for the parameters can be found at https://seaborn.pydata.org/generated/seaborn.jointplot.html

```python
# Recreate the chart above contrasting the total_bill amount and tips given
sns.jointplot(x=tips['total_bill'],y=tips['tip'],kind='reg',xlim=(0,60),ylim=(0,12),color='purple')
```

```python
# Modify one or more of the selected parameters to create a new version of this chart!
sns.jointplot(x=tips['total_bill'],y=tips['size'],kind='reg',xlim=(0,60),ylim=(0,12),color='red')
```
